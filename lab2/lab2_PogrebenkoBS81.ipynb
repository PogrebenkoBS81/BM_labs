Імпорт бібліотек, %matplotlib inline для піддтримки графіків у jupyter

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
y
```

2.Будування графіків.

    Функціональний метод.

```python
plt.plot(x,y)
plt.xlabel('Label X')
plt.ylabel('Label Y')
plt.title("X*X Title")
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'g')
```


    Об’єктно-орієнтований метод:

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y)
axes.set_xlabel('X Title')
axes.set_ylabel('Y Title')
axes.set_title('X*X Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.1,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_xlabel('X Title')
axes1.set_ylabel('Y Title')
axes1.set_title('X*X Title')

axes2.plot(y,x, 'g')
axes2.set_xlabel('X Title')
axes2.set_ylabel('Y Title')
axes2.set_title('X*X Title')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')

axes[1].plot(y,x)
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')
```

```python
fig, axes = plt.subplots(nrows=5, ncols=5)
plt.tight_layout()
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x**2, label='x^3')
axes.plot(x, x**3, label='sqrt(x)')
axes.legend()
```

```python
fig = plt.figure(figsize=(8,4), dpi = 200)
```

Кастомізація характеристик графіку:

```python
fig, axes = plt.subplots()
axes.plot(x, x*x, marker='o', markersize = 10, markerfacecolor='g', lw = 3, ls = ':')
```

Збереження графику у pdf.
