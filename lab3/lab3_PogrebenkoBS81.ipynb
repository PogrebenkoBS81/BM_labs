```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

Завантаження даних "tips" та вивід на екран перших 5 єлементів.

```python
tips = sns.load_dataset('tips')
tips.head()
```

Розподіл одновимірного набору спостережень.

```python
sns.distplot(tips['total_bill'],
                  kde_kws={"color": "b", "lw": 3, "label": "KDE"},
                  hist_kws={"histtype": "stepfilled", "linewidth": 1,
                            "alpha": 1, "color": "r"}, bins = 30)
```

Співставлення розподілу двох змінних та їх
взаємозв’язок за допомогою діаграми розсіювання з використанням jointplot

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind="hex",
                  marginal_kws=dict(bins=30, rug=True),
                    edgecolor="w", linewidth=1)
```

Попарні взаємозв’язки між змінними в повному наборі
даних.

```python
sns.pairplot(tips, diag_kind="kde", markers="+",
                 plot_kws=dict(s=50, edgecolor="b", linewidth=1),)
```

```python
sns.pairplot(tips, hue = 'sex')
```

Отримання сукупних даних по кількісним даним в
залежності від категоріальної змінної.

```python
sns.barplot(x = 'sex', y = 'total_bill', palette="Blues_d", data = tips)
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips, estimator = np.median)
```

Візуалізація розподілу категоріальних
даних за певною кількісною змінною.

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, palette = 'Set3')
```

Діаграма розсіювання, де одна за змінних
категоріальна за допомогою stripplot.

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips)
```

Графік взаємозв’язку (кореляції) між даними у вигляді
матриці.

```python
sns.heatmap(tips.corr(), annot = True)
```

Візуалізація зв’язоку між декількома змінними за допомогою FacetGrid.

```python
g = sns.FacetGrid(tips, col = 'time', row = 'smoker')
g = g.map(plt.hist, 'total_bill')
```
